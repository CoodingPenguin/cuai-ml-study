### 1. 조원의 정보를 가지고 DataFrame을 만들어 보세요.
#### 1.1. 조원들의 '이름', '나이', '학번', '전화번호'를 바탕으로 이를 열 이름으로 사용하여 DataFrame을 만들어보세요.

```python
col_name = ['이름', '나이', '학번', '전화번호']
mem_lst = [['박소현', '23', '20161305', '010-5389-5029'],
       ['박은우', '24', '20164583', '010-8131-2072'],
       ['정찬웅', '23', '20186663', '010-4113-8878'],
       ['황인택', '22', '20170000', '010-5559-3466']]

df_mem = pd.DataFrame(mem_lst, columns=col_name)
df_mem
```

#### 1.2. '나이' 열을 기존으로 오름차순으로 정렬해보세요.

```python
df_mem_sorted = df_mem.sort_values(by=['나이'])
df_mem_sorted
```

#### 1.3. '학번'에서 연도만 슬라이싱하여 이에 '4'를 더한 값을 새로운 열인 '졸업 연도'에 저장해보세요. 예를 들어, '학번'이 20189999라면, 2018만 추출하여 2022를 '졸업 연도' 열에 저장하면 됩니다.

```python
df_mem['졸업 연도'] = df_mem['학번'].apply(lambda x : int(x[:4]) + 4)
df_mem
```

#### 2.1. Go to Kaggle homepage and find ‘Google Play Store Apps’ dataset and download it. 

```python
app_df = pd.read_csv('./googleplaystore.csv')
app_df.drop(10472, axis=0, inplace=True)
app_df.head(7)
```

#### 2.2. Using method which is in Pandas, find and write column name which has ‘Not Null’ data. Fill Not Null data with this condition.
1. ‘Rating’ : mean()
2. ‘Content Rating’, ‘Type’ : Find largest count of this column and fill with it. (Hint : Using value_counts() method.)

```python
app_df.isna().sum()
```

```python
type_cnts = app_df['Type'].value_counts()
type_cnts
```

```python
content_cnts = app_df['Content Rating'].value_counts()
content_cnts
```

```python
app_df['Rating'] = app_df['Rating'].fillna(app_df['Rating'].mean())
app_df['Type'] = app_df['Type'].fillna('Free')
app_df['Content Rating'] = app_df['Content Rating'].fillna('Everyone')

app_df.isna().sum()
```

#### 2.3. Sorting ‘Size’ column in ascending order.

```python
app_df_sorted = app_df.sort_values(by=['Size'])
app_df_sorted.head(10)
```

#### 2.4. Drop ‘Current Ver’, ‘Android Ver’ columns.

```python
app_df.drop(['Current Ver', 'Android Ver'], axis=1, inplace=True)
app_df.head(10)
```

#### 2.5. ‘Rating_round’ column : The value of ‘Rating’ is rounded up. (반올림)

```python
app_df['Rating Round'] = app_df['Rating'].apply(lambda x: round(x))
app_df.head(10)
```

#### 2.6. ‘Rating_cat’ column : Using ‘apply lambda’ with this condition.
1. Using value of ‘Rating_round’.
2. And in the order of 1 to 5, fill it ‘Very Bad’, ‘Bad’, ‘So So’, ‘Good’, ‘Very Good’ order.

```python
def find_cat(score):
    if score < 1:
        return 'Very Bad'
    elif score >= 1 and score < 2:
        return 'Bad'
    elif score >= 2 and score < 3:
        return 'So So'
    elif score >= 3 and score < 4:
        return 'Good'
    elif score >= 4:
        return 'Very Good'

app_df['Rating Cat'] = app_df['Rating Round'].apply(lambda x: find_cat(x))
app_df.head(10)
```

#### 2.7. Using ‘groupby()’ method and indicate the mean rating by category.

```python
app_groupby = app_df.groupby('Category')['Rating'].mean()
app_groupby
```
